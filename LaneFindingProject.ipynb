```python
# 引入我自己的函数包
# import my packages
import LaneFindingPipline.LaneFinding as lf
```

```python
# 引入其他的包
# impot other packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
# 读取图片
#reading in an image
image = mpimg.imread('test_images/solidWhiteRight.jpg')

#printing out some stats and plotting
print('This image is:', type(image), 'with dimensions:', image.shape)
plt.imshow(image)
```

```python
# 灰度转换
# gray scale transform
image_gray = lf.gray_scale(image)

plt.imshow(image_gray)
```

```python
# 高斯滤波
# Applies a Gaussian Noise kernel
kernel_size = 3
image_gauss = lf.gaussian_blur(image_gray, kernel_size)

plt.imshow(image_gauss)
```

```python
# 边缘检测
# Applies the Canny transform
low_threshold = 50
high_threshold = 150
image_canny = lf.canny(image_gauss, low_threshold, high_threshold)

plt.imshow(image_canny)
```

```python
# 区域检测
# Applies an image mask.
imshape = image_canny.shape
vertices = np.array([[(0, imshape[0]),(450, 320), (550, 320), (imshape[1], imshape[0])]], dtype=np.int32)
image_interest = lf.region_of_interest(image_canny, vertices)

plt.imshow(image_interest)
```
