```python
# 引入我自己的函数包
# import my packages
import LaneFindingPipline.LaneFinding_lines as lf
```

```python
# 引入其他的包
# impot other packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
# 读取图片
#reading in an image
image = mpimg.imread('test_images/solidWhiteRight.jpg')

#printing out some stats and plotting
print('This image is:', type(image), 'with dimensions:', image.shape)
plt.imshow(image)
```

```python
# 灰度转换
# gray scale transform
image_gray = lf.gray_scale(image)

plt.imshow(image_gray)
```

```python
# 高斯滤波
# Applies a Gaussian Noise kernel
kernel_size = 3
image_gauss = lf.gaussian_blur(image_gray, kernel_size)

plt.imshow(image_gauss)
```

```python
# 边缘检测
# Applies the Canny transform
low_threshold = 50
high_threshold = 150
image_canny = lf.canny(image_gauss, low_threshold, high_threshold)

plt.imshow(image_canny)
```

```python
# 区域检测
# Applies an image mask.
imshape = image_canny.shape
vertices = np.array([[(0, imshape[0]),(460, 325), (520, 325), (imshape[1], imshape[0])]], dtype=np.int32)
image_interest = lf.region_of_interest(image_canny, vertices)

plt.imshow(image_interest)
```

```python
# 霍夫变换
# Hough transform
image_hough = lf.hough_lane(image_interest, rho = 1, theta = np.pi / 180, threshold = 15, min_line_len  = 40, max_line_gap = 20, y_min = 325, y_max = 540)

plt.imshow(image_hough)
```

```python
image_res = lf.weighted_img(image_hough, image, 0.8, 1, 0)

plt.imshow(image_res)
```

```python
# 图片处理管道
# image pip line
image_res = lf.process_image_lane(image)

plt.imshow(image_res)
mpimg.imsave('test_images_output/solidWhiteRight_lane.jpg', image_res)
```
